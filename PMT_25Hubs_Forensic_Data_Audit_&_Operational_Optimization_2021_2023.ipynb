```python
import pandas as pd
import numpy as np
import os

folder_name = 'PMT_Professional_Portfolio'
os.makedirs(folder_name, exist_ok=True)

hubs = ["ENUGU NORTH", "OSHA MAIN", "OSHA OWERRI RD/ASABA", "UMUAHIA", "BENIN", "YENEGOA", "NNEWI", "AFIKPO", "UYO", "WARRI", "IKOM", "CALABAR BEDWELL", "ABA", "LAGOS (MAZAMAZA)", "ABUJA (UTAKO)", "PHC", "KADUNA", "KANO", "IBADAN", "JOS", "MAKURDI", "BAUCHI", "AKURE", "ONITSHA UPPER IWEKA", "OWERRI"]
profiles = {"LAGOS (MAZAMAZA)": 1.45, "PHC": 1.35, "ABUJA (UTAKO)": 1.35, "YENEGOA": 0.5, "IKOM": 0.5, "JOS": 0.5, "CALABAR BEDWELL": 0.6}
dates = pd.date_range(start="2021-01-01", end="2023-12-31", freq='MS')

records_ops = []
for hub in hubs:
    perf = profiles.get(hub, 1.0)
    for date in dates:
        gross = np.random.uniform(45000000, 65000000) * perf
        ops_70 = gross * 0.70
        leak = np.random.uniform(1.2, 1.6) if perf < 0.9 else 1.0
        fuel = (gross * np.random.uniform(0.32, 0.40)) * leak
        maint = (np.random.uniform(2000000, 4500000) * leak)
        loans = np.random.choice([0, 7000000], p=[0.85, 0.15])
        net_inc = ops_70 - (fuel + 4500000 + maint + 1500000 + loans)
        records_ops.append({'Date': date, 'Hub_Name': hub, 'Data_Source': 'Company_Data_Warehouse', 'Operating_Income_70': round(ops_70, 2), 'Fuel_Costs': round(fuel, 2), 'Maintenance_MTS': round(maint, 2), 'Director_Remuneration': 1500000, 'Intervention_Loans': round(loans, 2), 'Net_Income_Final': round(net_inc, 2)})

df_ops = pd.DataFrame(records_ops)
df_ops.to_csv(f'{folder_name}/1_Monthly_Operational_NetIncome_36M.csv', index=False)
print("SUCCESS: 1_Monthly_Operational_NetIncome_36M.csv generated.")
```

```python
df_summary = df_ops.groupby('Hub_Name').sum(numeric_only=True).reset_index()
df_summary.to_csv(f'{folder_name}/2_ThreeYear_Operational_Performance_Summary.csv', index=False)
print("SUCCESS: 2_ThreeYear_Operational_Performance_Summary.csv generated.")
```

```python
import pandas as pd

# List of our 6 original files
original_files = [
    '1_Monthly_Operational_NetIncome_36M.csv',
    '2_ThreeYear_Operational_Performance_Summary.csv',
    '3_Monthly_Repayment_Schedule_36M.csv',
    '4_Half_Yearly_Repayment_Schedule.csv',
    '5_ThreeYear_Total_Repayment_Status.csv',
    '6_Forensic_Analyst_Audit_Report.csv'
]

print("--- Starting Master Cleaning Process ---")

for file_name in original_files:
    file_path = f'PMT_Professional_Portfolio/{file_name}'

    # 1. Load File
    df = pd.read_csv(file_path)

    # 2. Convert Dates (if the file has a Date column)
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'])

    # 3. Handle Nulls (Filling empty numeric cells with 0)
    df = df.fillna(0)

    # 4. Remove Duplicates
    df = df.drop_duplicates()

    # 5. Round Financials to 2 Decimal Places
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    df[numeric_cols] = df[numeric_cols].round(2)

    # 6. Save as a CLEANED version
    new_name = f"CLEANED_{file_name}"
    df.to_csv(f'PMT_Professional_Portfolio/{new_name}', index=False)

    print(f"Verified & Cleaned: {new_name}")

print("\n--- All 6 Files are now Vetted and Ready for Dashboard Import ---")
```

```python
import pandas as pd
import numpy as np

# Create the 36-month timeline
months = pd.date_range(start='2021-01-01', end='2023-12-01', freq='MS')
hubs = [
    'LAGOS (MAZAMAZA)', 'PHC', 'ABUJA (UTAKO)', 'ENUGU NORTH', 'WARRI',
    'AKURE', 'KANO', 'KADUNA', 'ABA', 'MAKURDI', 'BAUCHI', 'IBADAN',
    'UYO', 'OWERRI', 'NNEWI', 'OSHA MAIN', 'BENIN', 'YENEGOA', 'IKOM',
    'JOS', 'CALABAR BEDWELL'
]

data = []
for month in months:
    for hub in hubs:
        # Standard Income
        income = np.random.uniform(8000000, 15000000)

        # High Risk Logic: Group D has higher costs and bigger loans
        if hub in ['YENEGOA', 'IKOM', 'JOS', 'CALABAR BEDWELL']:
            costs = income * np.random.uniform(0.85, 0.95) # Barely breaking even
            loan = np.random.uniform(800000, 1500000)      # High debt burden
        else:
            costs = income * np.random.uniform(0.60, 0.75) # Healthy margins
            loan = np.random.uniform(100000, 400000)       # Normal operations

        data.append({
            'Date': month,
            'Hub': hub,
            'Monthly Revenue (₦)': round(income, 2),
            'Operating Costs (₦)': round(costs, 2),
            'Intervention Loan (₦)': round(loan, 2),
            'Net Monthly Balance (₦)': round(income - costs - loan, 2)
        })

df_monthly_master = pd.DataFrame(data)
df_monthly_master.to_csv('PMT_36Month_Master_Logs.csv', index=False)
print("Success: PMT_36Month_Master_Logs.csv is ready for download!")
```
